```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
fish = cv2.imread('images/fish.bmp', cv2.IMREAD_COLOR)
submarine = cv2.imread('images/submarine.bmp', cv2.IMREAD_COLOR)
```

```python
plt.imshow(cv2.cvtColor(fish, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()

plt.imshow(cv2.cvtColor(submarine, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
# ensure same sizes
```

```python
# low pass einstein, high pass monroe
submarine_low_passed = cv2.GaussianBlur(submarine, (13, 13), 0)
fish_high_passed = fish - cv2.GaussianBlur(fish, (13, 13), 0)
```

```python
plt.imshow(cv2.cvtColor(submarine_low_passed, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()

plt.imshow(cv2.cvtColor(fish_high_passed, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
hybrid = submarine_low_passed + fish_high_passed
# show all 3 images side by side
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(submarine, cv2.COLOR_BGR2RGB))
plt.title('Submarine')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(cv2.cvtColor(fish, cv2.COLOR_BGR2RGB))
plt.title('Fish')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(cv2.cvtColor(hybrid, cv2.COLOR_BGR2RGB))
plt.title('Hybrid')
plt.axis('off')

plt.show()
```
